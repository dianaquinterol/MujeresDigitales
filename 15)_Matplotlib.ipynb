```python
x = np.arange(0,10)
y = np.arange(0,10)

median = 5

plt.plot(x,y, label="Linear")
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(loc='upper center')
plt.fill_between(x,y,median, alpha=0.25)
```

```python
x = np.arange(0,10)
y = np.arange(0,10)

median = 3

plt.plot(x,y, label="Linear")
plt.ylabel('Y')
plt.xlabel('X')
plt.legend(loc='upper center')
plt.fill_between(x,y,median, where=(y > median), interpolate=True, alpha=0.25)
plt.fill_between(x,y,median, where=(y <= median), interpolate=True, color='red', alpha=0.25)
```

```python
x = np.arange(0,10)
y =  np.arange(0,10)
y2 = [0,1,1,2,3,6,7,7,5,6]

plt.plot(x,y, label="Linear")
plt.plot(x,y2, label="Linear 2")
plt.ylabel('Y')
plt.xlabel('X')
plt.fill_between(x,y,y2, where=(y > y2), interpolate=True, alpha=0.25, label='Above y2')
plt.fill_between(x,y,y2, where=(y <= y2), interpolate=True, color='red', alpha=0.25, label='Below y2')
plt.legend(loc='upper left')
```

```python
from matplotlib.animation import FuncAnimation

def anim(i):
    x_v  = np.arange(0,10)
    y_v  = np.random.randint(0,10, size=(1,10))[0]
    y_v2 = np.random.randint(0,10, size=(1,10))[0]
    
    plt.cla() 
    plt.plot(x_v, y_v, label="CH1")
    plt.plot(x_v,y_v2, label="CH2")
    
    plt.legend(loc='upper left')
    plt.tight_layout()
    
    
an = FuncAnimation(plt.gcf(), anim, interval=10000)

```
